```python
import torch.nn as nn
import torch.nn.functional as F

class CNN(nn.Module):
    def __init__(self, input_dim=1, output_dim=99*2):
        super().__init__()
        
        self.without_start_goal = nn.Sequential(
            nn.Conv2d(input_dim, 64, kernel_size=4, stride=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            
            nn.AvgPool2d(kernel_size=4, stride=3),
            
            nn.Conv2d(64, 32, kernel_size=4, stride=1),
            nn.BatchNorm2d(32),
            nn.ReLU(inplace=True),
            
            nn.AvgPool2d(kernel_size=4, stride=3),
            
            nn.Conv2d(32, 32, kernel_size=4, stride=1),
            nn.BatchNorm2d(32),
            nn.ReLU(inplace=True),
            
            nn.AvgPool2d(kernel_size=4, stride=3),
        )
        
        self.project_start_and_goal = nn.Sequential(
            nn.Linear(4,8),
            nn.Linear(8,16),
            nn.Dropout(p=0.2),
            nn.Linear(16,32),
            nn.Dropout(p=0.2),
            nn.Linear(32,64),
        )
        
        self.map_weight = 3
        self.middle_dim = 128 + 64
        self.fc1 = nn.Linear(self.middle_dim, 128)
        self.fc2 = nn.Linear(128, output_dim)
    
    def forward(self, map_info, start_and_goal, verbose=False):
        map_info = self.without_start_goal(map_info)
        map_info = (map_info.reshape(1, -1))*self.map_weight
        if verbose:
            print("After first set of conv layers: ", map_info.size())
        
        start_and_goal = self.project_start_and_goal(start_and_goal)
        if verbose:
            print("Projected start and goal:", start_and_goal.size())
        
        output = torch.cat((map_info, start_and_goal), 1)
        if verbose:
            print("Concatenated: ", output.size())
                
        return self.fc2(self.fc1(output))
```

```python
def prepare_training_data(trial, include_sdf=False):
    map_tensor = trial["map_tensor"].unsqueeze(1)
    sdf_tensor = trial["sdf_data"].unsqueeze(1)
    expert_traj = trial["expert_trajectory"].to(cuda0)
    expert_traj = expert_traj[:][0:2][:]
    label_traj = torch.cat((expert_traj[:][0][1:-1], expert_traj[1][1:-1]), 1).to(cuda0)
    start_x, start_y = expert_traj[0][0], expert_traj[0][-1]
    goal_x, goal_y = expert_traj[1][0], expert_traj[1][-1]
    start_and_goal = torch.tensor([[start_x, start_y, goal_x, goal_y]]).to(cuda0)
    # map_tensor = (map_tensor.expand((1,1,128,128))).double().to(cuda0)
    
    if include_sdf:
        map_tensor = torch.cat((map_tensor, sdf_tensor), 1).expand((1,2,128,128)).double().to(cuda0)
    else:
        map_tensor = (map_tensor.expand((1,1,128,128))).double().to(cuda0)
    
    return map_tensor, start_and_goal, label_traj

def loss_func(feat1, feat2):
    # maximize average magnitude of cosine similarity
    loss = nn.MSELoss()
    return loss(feat1, feat2)
```

```python
net = CNN(input_dim=2, output_dim=198).to(cuda0)
ex_trial = next(iter(train_dl))
# print(ex_trial["expert_trajectory"])
map_tensor, start_and_goal, label_traj = prepare_training_data(ex_trial, include_sdf=True)
# print(start_and_goal)
# rint(label_traj.size())
out = net(map_tensor, start_and_goal, verbose=True)
```

```python
# Training loop
from torch.optim.lr_scheduler import ExponentialLR
include_sdf = False

net = CNN(input_dim=1+include_sdf, output_dim=198).to(cuda0)
optimizer = torch.optim.Adam(net.parameters(), lr = 0.001)
scheduler = ExponentialLR(optimizer, gamma=0.9)
net.train()
epochs = 50

for epoch in range(epochs):
    epoch_losses = []
    for trial in train_dl:
        map_tensor, start_and_goal, label_traj = prepare_training_data(trial, include_sdf=include_sdf)
        net.zero_grad()
        traj = net(map_tensor, start_and_goal)
        loss = loss_func(traj, label_traj)
        loss.backward()
        optimizer.step()
        epoch_losses.append(loss)
        # print(loss)
    
    loss = sum(epoch_losses)/len(epoch_losses)
    scheduler.step()
    print("Done training {}/{}, MSE_loss = {:.4f}".format(epoch+1, epochs, loss))
```
